# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

From the Centers for Medicare & Medicaid Services (CMS):

    The Hospital Readmissions Reduction Program (HRRP) is a Medicare value-based purchasing program that reduces payments to hospitals with excess readmissions. The program supports the national goal of improving healthcare for Americans by linking payment to the quality of hospital care.

    Section 3025 of the Affordable Care Act requires the Secretary of the Department of Health and Human Services (HHS) to establish HRRP and reduce payments to Inpatient Prospective Payment System (IPPS) hospitals for excess readmissions beginning October 1, 2012 (i.e., Federal Fiscal Year [FY] 2013). Additionally, the 21st Century Cures Act requires CMS to assess penalties based on a hospital’s performance relative to other hospitals with a similar proportion of patients who are dually eligible for Medicare and full-benefit Medicaid beginning in FY 2019. The legislation requires estimated payments under the stratified methodology to equal payments under the non-stratified methodology (i.e., the methodology from FY 2013 to FY 2018) to maintain budget neutrality. The payment reduction is capped at 3% (i.e., payment adjustment factor of 0.97). Payment reductions are applied to all Medicare FFS base operating DRG payments between October 1, 2018 through September 30, 2019. CMS uses excess readmission ratios (ERR) to measure performance for each of the six conditions/procedures in the program:

    Acute Myocardial Infarction (AMI)
    Chronic Obstructive Pulmonary Disease (COPD)
    Heart Failure (HF)
    Pneumonia (PN)
    Coronary Artery Bypass Graft (CABG) Surgery
    Elective Primary Total Hip Arthroplasty and/or Total Knee Arthroplasty (HIP-KNEE)

#### 4. Discuss statistical significance and practical significance. Do they differ here? How does this change your recommendation to the client?#### 

**Statistical Significance** quantifies the likelihood that the mean differences observed in the sample have occurred due to sampling error. The one caviot is that statistical significance can be found in large samples sizes despite seemingly insignificant differences. 
**Practical Significance** accounts accounts for "effect size", meaning, it assesses if the differences are large enough in value in a "real life applicable" sense. Assessing Practical Significance allows analysts to move beyond the simplistic, 'Does it work or not?' to the far more sophisticated, 'How well does it work in a range of contexts?'

For this study, it *is statistical signicance* that the hospital's number of discharges is correlated with the hospital's excess readmission ration (p-value = 1.22 e-25). However, the finding is *not practical significance* because the effective correlation coefficient (r = -0.097) is considered very small and therefore is insignificant.

Since the finding of this study was found to not be practically significant, this changes the recommendation to the client because the assumption that excess readmission can be attributed to hospital size cannot be accepted in a practical sense. Therefore, advising that a hospital's to upgraded resource allocation and/or consolidate their facilities in order to reduce readmission rates would no longer relevant. Further analyses would need to be performed prior to advising the client to enforce any policy changes.
